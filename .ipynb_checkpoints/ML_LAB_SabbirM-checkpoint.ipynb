## Problem 1: Dataset Import & Cleaning<br>
Check **"Profit"** and **"Sales"** in the dataset, convert these two columns to numeric type.

## Problem 2: <br>
- Retailers that depend on seasonal shoppers have a particularly challenging job when it comes to inventory management. Your manager is making plans for next year's inventory.
- He wants you to answer the following questions:
    1. Is there any seasonal trend of inventory in the company?
    2. Is the seasonal trend the same for different categories?
- ***Hint:*** For each order, it has an attribute called `Quantity` that indicates the number of product in the order. If an order contains more than one product, there will be multiple observations of the same order.

```python
labDFcopy['Order.Date'] = pd.to_datetime(labDFcopy['Order.Date'], format = '%m/%d/%y')
labDFcopy['Ship.Date'] = pd.to_datetime(labDFcopy['Ship.Date'], format = '%m/%d/%y')

labDFcopy['OrderMonth'] = [labDFcopy['Order.Date'][i].month for i in range(len(labDFcopy['Order.Date']))]
labDFcopy['OrderYear'] = [labDFcopy['Order.Date'][i].year for i in range(len(labDFcopy['Order.Date']))]
```

```python
lab2a = labDFcopy.groupby(labDFcopy['OrderMonth'])
SeasonalInventoryTrend = lab2a['Quantity'].agg(['sum']).reset_index()
```

```python
plt.figure(figsize=(12,5))
plt.bar(SeasonalInventoryTrend['OrderMonth'], height = SeasonalInventoryTrend['sum'])
plt.xlabel('Month of the Year')
plt.ylabel('Amount of items (inventory) Sold')
plt.title('Seasonal Trend of Inventory Sold')
plt.show()
```

```python
lab2b = labDFcopy.groupby(['OrderMonth','Category'])
SeasonalCategoryTrend = lab2b['Quantity'].agg(['sum']).reset_index()
```

```python
plt.figure(figsize=(12,5))
g = sns.barplot(x='OrderMonth', y='sum', hue='Category', data=SeasonalCategoryTrend)
plt.xlabel('Month of the Year')
plt.ylabel('Amount of items (inventory) Sold')
plt.title('Seasonal Trend of Inventory Sold by Category')
plt.show()
```

## Problem 3:<br>
- Your manager required you to give a brief report (**Plots + Interpretations**) on returned orders.
	1. How much profit did we lose due to returns each year?
	2. How many customer returned more than once? more than 5 times?
	3. Which regions are more likely to return orders?
	4. Which categories (sub-categories) of products are more likely to be returned?
- ***Hint:*** Merge the **Returns** dataframe with the **Orders** dataframe using `Order.ID`.

*Renaming column name in Returns data frame and merging it with Orders data frame*

```python
labDF2copy = labDF2copy.rename(columns={'Order ID':'Order.ID'})
labDFfull = pd.merge(labDFcopy, labDF2copy, on = 'Order.ID', how = 'outer')
returnedDF = labDFfull.loc[labDFfull.Returned == 'Yes',:]
```

1. Profits lost due to returns, by year

```python
lab3_1 = returnedDF.groupby(['OrderYear'])
yearlyReturnLoss = lab3_1.Profit.agg(['sum']).reset_index()
yearlyReturnLoss
```

2. How many customer returned more than once? more than 5 times?

```python
lab3_2 = returnedDF.groupby(['Customer.ID'])
OneTimeReturn = lab3_2.Returned.agg(['count']).sort_values('count', ascending=False).reset_index()
FiveTimeReturn = OneTimeReturn.loc[OneTimeReturn['count']>5]
print('Number of customers who returned more than once: ',len(OneTimeReturn))
print('Number of customers who returned more than 5 times: ',len(FiveTimeReturn))
```

3. Which regions are more likely to return orders?

```python
lab3_3 = returnedDF.groupby(['Region_y'])
RegionReturn = lab3_3.Returned.agg(['count']).sort_values('count', ascending=False).reset_index()
RegionReturn
```

4. Which categories (sub-categories) of products are more likely to be returned?

```python
lab3_4 = returnedDF.groupby(['Sub.Category'])
SubCategoryReturn = lab3_4.Returned.agg(['count']).sort_values('count', ascending=False).reset_index()
SubCategoryReturn
```

#### Step 2:

- He wants you to generate a feature which can measure how long it takes the company to process each order.
- ***Hint:*** Process.Time = Ship.Date - Order.Date

#### Step 3:
- If a product has been returned before, it may be returned again. 
- Let us generate a feature indictes how many times the product has been returned before.
- If it never got returned, we just impute using 0.
- ***Hint:*** Group by different Product.ID

**Step 1:**

```python
labDFfull['Returned'] = labDFfull['Returned'].map(lambda x: 1 if x=='Yes' else 0)
```

**Step 2:**

```python
labDFfull['Process.Time'] = labDFfull['Ship.Date'] - labDFfull['Order.Date']
```

**Step 3:**

```python
lab4_3 = labDFfull.groupby(['Product.ID'])
ProductReturn = lab4_3.Returned.agg(['sum']).sort_values('sum', ascending=False).reset_index()
```
